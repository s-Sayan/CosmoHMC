```python
import healpy as hp
import numpy as np
import cosmohmc.almncl.utils as utils
import camb
import os.path as op
import matplotlib.pyplot as plt
```

```python
cambinifile = 'planck_2018_acc'

pars = camb.read_ini(op.join(op.dirname('/Users/sayan/CMB_WORK/CAMB-1.1.3/inifiles'),  'inifiles', cambinifile + '.ini'))
results = camb.get_results(pars)
```

```python
lmax = 16
cpp_fid = results.get_lens_potential_cls(lmax=lmax, raw_cl=True).T[0]

camb_cls = results.get_unlensed_scalar_cls(CMB_unit='muK', raw_cl=True, lmax=lmax).T
cls_unl_fid = {'tt':camb_cls[0], 'ee':camb_cls[1], 'bb':camb_cls[2], 'te':camb_cls[3], 'pp':cpp_fid}
```

```python
alms_t = hp.synalm(cls_unl_fid['tt'])
#alms_t = hp.synalm(cpp_fid)
```

```python
mapt = hp.alm2map(alms_t, nside=16)
```

```python
hp.mollview(map=mapt,
            fig=None,
            rot=None, 
            coord=None, 
            unit=r'microK_CMB', 
            min=None, 
            max=None, 
            cbar=True, 
            cmap='YlGnBu', 
            notext=False, 
            norm=None, 
            hold=False, 
            margins=None, 
            sub=None, 
            return_projected_map=False)
```

```python
almr, almi = utils.convert_alm_to_2d(alms_t, lmax)
```

```python
alms_rec = utils.reconstruct_alm_from_2d(almr, almi, lmax)
```
